```python
# Importing necessary modules
import matplotlib.pyplot as plt
import numpy as np
```

```python
# loading data
def load_data(filename='parameter_influence_data.npz'):
    """Load dictionary from NumPy file"""
    loaded = np.load(filename)
    
   
    data = {}
    for key in loaded.files:
        main_key, sub_key = key.rsplit('_', 1)
        if main_key not in data:
            data[main_key] = {}
        data[main_key][sub_key] = loaded[key].tolist()
    
    print(f"Data loaded from {filename}")
    return data
```

```python
Data = load_data()
```

```python
# Data

sigma_H_1 = Data['sigma_H']['1'] 
sigma_H_2 = Data['sigma_H']['2']
sigma_H_3 = Data['sigma_H']['3']
sigma_H_x = Data['sigma_H']['x']

sigma_v_1 = Data['sigma_v']['1']
sigma_v_2 = Data['sigma_v']['2']  
sigma_v_3 = Data['sigma_v']['3'] 
sigma_v_x = Data['sigma_v']['x'] 

sigma_h_1 = Data['sigma_h']['1'] 
sigma_h_2 = Data['sigma_h']['2'] 
sigma_h_3 = Data['sigma_h']['3'] 
sigma_h_x = Data['sigma_h']['x'] 

sigma_ovp_1 = Data['sigma_ovp']['1'] 
sigma_ovp_2 = Data['sigma_ovp']['2'] 
sigma_ovp_3 = Data['sigma_ovp']['3'] 
sigma_ovp_x = Data['sigma_ovp']['x'] 
```

```python


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(9, 6))

# Defining colors
color1 = 'tab:red'
color2 = 'blue'
color3 = 'tab:green'

# Subplot 1: Sigma H 
ax1.set_xlabel(r'$\sigma_H$', fontsize=16)
#ax1.set_ylabel('Potency', color=color1)
line1_1 = ax1.plot(sigma_H_x, sigma_H_1, color=color1, linewidth=3)
line1_2 = ax1.plot(sigma_H_x, sigma_H_2, color=color2, linewidth=3)

ax1.tick_params(axis='y', labelcolor='black')
#ax1.grid(True, alpha=0.3)
ax1.set_yticks([0, 30000, 60000, 90000])
ax1.set_xticks([200, 250, 300, 350])
ax1.tick_params(axis='both', labelsize=14, length=8, width=2)

for spine in ax1.spines.values():
    spine.set_linewidth(2)  
    spine.set_color('black')  


# Create second y-axis for subplot 1
ax1_twin = ax1.twinx()
#ax1_twin.set_ylabel('Potency', color=color3)
line1_3 = ax1_twin.plot(sigma_H_x, sigma_H_3, color=color3, linewidth=3)

ax1_twin.tick_params(axis='y', labelcolor=color3)
ax1_twin.set_yticks([0, 500, 1000, 1500])
ax1_twin.set_xticks([200, 250, 300, 350])
ax1_twin.tick_params(axis='both', labelsize=14, length=8, width=2)

# Subplot 2: Sigma V
ax2.set_xlabel(r'$\sigma_v$', fontsize=16)
#ax2.set_ylabel('Potency', color=color1)
line2_1 = ax2.plot(sigma_v_x, sigma_v_1, color=color1, linewidth=3)
line2_2 = ax2.plot(sigma_v_x, sigma_v_2, color=color2, linewidth=3)
ax2.tick_params(axis='y', labelcolor='black')
#ax2.grid(True, alpha=0.3)
ax2.set_yticks([2000, 8000, 14000, 20000])
ax2.set_xticks([140, 180, 220, 260])
ax2.tick_params(axis='both', labelsize=14, length=8, width=2)

for spine in ax2.spines.values():
    spine.set_linewidth(2)  
    spine.set_color('black')  


##################################################################
#subplot 2
ax2_twin = ax2.twinx()
#ax2_twin.set_ylabel('Potency', color=color3)
line2_3 = ax2_twin.plot(sigma_v_x, sigma_v_3, color=color3, linewidth=3)
ax2_twin.tick_params(axis='y', labelcolor=color3)
ax2_twin.set_yticks([0, 400, 800, 1200])
ax2_twin.set_xticks([140, 180, 220, 260])
ax2_twin.tick_params(axis='both', labelsize=14, length=8, width=2)

# Subplot 3: Sigma H 
ax3.set_xlabel(r'$\sigma_h$', fontsize=16)
ax3.set_ylabel('Potency [m$^3$]', color=color1, fontsize=16)
line3_1 = ax3.plot(sigma_h_x, sigma_h_1, color=color1, linewidth=3)
line3_2 = ax3.plot(sigma_h_x, sigma_h_2, color=color2, linewidth=3)
ax3.tick_params(axis='y', labelcolor='black')
#ax3.grid(True, alpha=0.3)
ax3.set_yticks([0, 4000, 8000, 12000])
ax3.set_xticks([60, 100, 140, 180])
ax3.tick_params(axis='both', labelsize=14, length=8, width=2)

for spine in ax3.spines.values():
    spine.set_linewidth(2)  
    spine.set_color('black')  


#################################################################
#subplot 3
ax3_twin = ax3.twinx()
ax3_twin.set_ylabel('Potency/1e3 [m$^3$]', color=color3, fontsize=16)
line3_3 = ax3_twin.plot(sigma_h_x, sigma_h_3, color=color3, linewidth=3)
ax3_twin.tick_params(axis='y', labelcolor=color3)
ax3_twin.set_yticks([0, 500, 1000, 1500])
ax3_twin.set_xticks([60, 100, 140, 180])

ax3_twin.tick_params(axis='both', labelsize=14, length=8, width=2)

# Subplot 4: Sigma OVP
ax4.set_xlabel(r'$\sigma_p$', fontsize=16)
#ax4.set_ylabel('Potency', color=color1)
line4_1 = ax4.plot(sigma_ovp_x, sigma_ovp_1, color=color1, linewidth=3)
line4_2 = ax4.plot(sigma_ovp_x, sigma_ovp_2, color=color2, linewidth=3)
ax4.tick_params(axis='y', labelcolor='black')
#ax4.grid(True, alpha=0.3)
ax4.set_yticks([2000, 6000, 10000, 14000])
ax4.set_xticks([60, 90, 120, 150])
ax4.tick_params(axis='both', labelsize=14, length=8, width=2)

for spine in ax4.spines.values():
    spine.set_linewidth(2)  
    spine.set_color('black')  


#####################################################################
#subplot 4
ax4_twin = ax4.twinx()
#ax4_twin.set_ylabel('Potency', color=color3)
line4_3 = ax4_twin.plot(sigma_ovp_x, sigma_ovp_3, color=color3, linewidth=3)
ax4_twin.tick_params(axis='y', labelcolor=color3)
ax4_twin.set_yticks([0, 4, 8, 12])
ax4_twin.set_xticks([60, 90, 120, 150])

ax4_twin.tick_params(axis='both', labelsize=14, length=8, width=2)

# Adjust layout to prevent overlapping
plt.tight_layout()


plt.savefig("parameter_influence.png",dpi=400)
```
