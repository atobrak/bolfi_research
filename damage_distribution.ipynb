```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the saved data
edge_length_data = np.load('edge_length_data.npy')
damage_data = np.load('damage_data.npy')
```

```python
# Create the figure with subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 4))

# Plot (a) - Approximate element edge length
ax1.hist(edge_length_data, bins=25, color='blue', edgecolor='black', linewidth=0.001)
ax1.axvline(x=45, color='red', linestyle='--', linewidth=2)  # Reference line
ax1.set_xlabel('Approximate element edge length [m]')
ax1.set_ylabel('Frequency')
ax1.set_xlim(20, 80)
ax1.set_ylim(0, 2200)

ax1.text(0.01, 0.85, '(a)', transform=ax1.transAxes, fontsize=12)
#ax1.grid(True, alpha=0.3)

# Plot (b) - Element damage value
ax2.hist(damage_data, bins=35, color='blue', edgecolor='black', linewidth=0.001)
ax2.axvline(x=0.095, color='red', linestyle='--', linewidth=2)  # Reference line
ax2.set_xlabel('Element damage value [-]')
ax2.set_ylabel('Frequency')
ax2.set_xlim(0.04, 0.20)
ax2.set_ylim(0, 3200)
ax2.text(0.01, 0.85, '(b)', transform=ax2.transAxes, fontsize=12)
#ax2.grid(True, alpha=0.3)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.savefig("mesh_information.png", dpi=400)
#plt.show()
```
